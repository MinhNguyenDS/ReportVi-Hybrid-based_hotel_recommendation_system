```python
import pandas as pd
from unidecode import unidecode
import numpy as np
import re
```

```python
df_booking = pd.read_csv('/content/drive/MyDrive/DS300_DoAn/data/data_booking.csv')
df_travel = pd.read_csv('/content/drive/MyDrive/DS300_DoAn/data/data_travel.csv')
df_ivivu = pd.read_csv('/content/drive/MyDrive/DS300_DoAn/data/data_ivivu.csv')
```

```python
def remove_accents(text):
    return unidecode(text)

df_booking.CustomerName = df_booking.CustomerName.apply(remove_accents)
```

```python
unique_names = df_booking[df_booking.CustomerName != 'Khong ten'].CustomerName.unique()
```

```python
unique_names = ['Nguyen', 'Tran', 'Le', 'Pham', 'Huynh']
```

```python
for index, row in df_booking.iterrows():
  if row.CustomerName == 'Khong ten':
    new_name = np.random.choice(unique_names)
    df_booking.at[index, 'CustomerName'] = new_name
```

```python
def text_preprocess(document):
  document = re.sub(r'[()*]','', document)
  document = re.sub(r'[_-]',' ', document)
  document = re.sub(r'(.)\1+', r'\1', document)
  document = re.sub(r'\d', '', document)
  document = document.replace('@gmail', '')

  return document
```

```python
characters = ['T', 'H', 'N', 'M', 'K', 'L', 'C', 'P', 'D', 'B']
weights = [0.3, 0.3, 0.08, 0.08, 0.08, 0.075, 0.025, 0.02, 0.02, 0.02]

def covert_name4booking(document):
    random_chars = np.random.choice(characters, size=np.random.randint(2, 4), p=weights)

    return document + ' ' + '. '.join(random_chars) + '.'
```

```python
df_booking.CustomerName = df_booking.CustomerName.apply(text_preprocess)
df_ivivu.CustomerName = df_ivivu.CustomerName.apply(text_preprocess)
df_travel.CustomerName = df_travel.CustomerName.apply(text_preprocess)
df_booking.CustomerName = df_booking.CustomerName.apply(covert_name4booking)
```

```python
df_general = pd.concat([df_booking, df_travel, df_ivivu])

df_general = df_general.rename(columns={'Ratting': 'Rating'})
df_general = df_general.rename(columns={'Address': 'Location'})
```

```python
df_general.Date = pd.to_datetime(df_general['Date'])
df_general = df_general.sort_values('Date')
```
