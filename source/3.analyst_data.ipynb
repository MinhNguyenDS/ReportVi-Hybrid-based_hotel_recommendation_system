```python
def normalize_value_counts(vector_counts):
    # Tạo một bản sao của mảng ban đầu
    mapping = {val: idx for idx, val in enumerate(sorted(vector_counts.index))}
    vector_counts.index = vector_counts.index.map(mapping)
    return vector_counts
```

```python
# Tạo bảng tổng hợp số lượng đánh giá theo khách sạn
plt.figure(figsize=(8, 6))
user_counts = df_history['IDuser'].value_counts()
user_counts = user_counts[user_counts >= 10]
user_counts = normalize_value_counts(user_counts)
# user_counts = normalize_value_counts(df_history['IDuser'].value_counts().head(500))
plt.bar(user_counts.index, user_counts.values)
plt.title('Số lượng đánh giá của các khách hàng')
plt.xlabel('ID khách hàng')
plt.ylabel('Số lượng đánh giá')

plt.show()
```

```python
df_history['Rating'] = df_history['Rating'].map(lambda x: round(x)).astype(int)
rating_distribution = df_history['Rating'].value_counts()
rating_distribution = rating_distribution.sort_index(ascending=False)

# Gộp các phần trăm nhỏ hơn 3% thành một nhóm 'Other'
threshold = 0.03
small_percentages = rating_distribution[rating_distribution / rating_distribution.sum() < threshold]
rating_distribution['Other'] = small_percentages.sum()
rating_distribution = rating_distribution[rating_distribution / rating_distribution.sum() >= threshold]

# Vẽ biểu đồ
plt.figure(figsize=(6, 6))
plt.pie(rating_distribution.values, labels=rating_distribution.index, autopct='%1.1f%%', startangle=90)
plt.title('Phân phối đánh giá của khách sạn')

plt.show()
```

```python
# Tạo bảng tổng hợp tỷ lệ địa điểm chứa khách sạn
location_percentage = df_hotel['Location'].value_counts(normalize=True) * 100

# Vẽ biểu đồ tròn
plt.pie(location_percentage.values, labels=location_percentage.index, autopct='%1.1f%%')

# Đặt tiêu đề cho biểu đồ
plt.title('Tỷ lệ các địa điểm chứa khách sạn')

# Hiển thị biểu đồ
plt.show()
```

```python
# Sắp xếp giá trị của cột Price
price_sort = df_info_hotel.sort_values('Price').Price

# Xác định các ngưỡng giá trị cho các mức thấp, trung bình, cao
low_threshold = round(price_sort[price_sort.index[int(len(price_sort)/3) - np.random.randint(0, 5)]], -3)
high_threshold = round(price_sort[price_sort.index[int(len(price_sort)/3)*2 + np.random.randint(0, 5)]], -3)

# Hàm để mã hóa giá trị
def encode_price_category(price):
    if price <= low_threshold:
        return 'Thấp'
    elif price <= high_threshold:
        return 'Trung bình'
    else:
        return 'Cao'

# Tạo cột mới chứa các mức giá được mã hóa
df_info_hotel['Price_Category'] = df_info_hotel['Price'].apply(encode_price_category)
```
